```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Medium', 'High']
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right = False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
# Examine the distribution of unique categories in categorical columns
categorical_cols = ['BRId', 'IAId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation','Fee Structure','Loyalty Classification', 'Properties Owned', 'Risk Weighting','Income Band']

for col in categorical_cols:
    print(f'Value Counts for "{col}":')
    display(df[col].value_counts())
```

## Univariant Analysis

```python
for i, predictor in enumerate(df[['BRId', 'IAId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation','Fee Structure','Loyalty Classification', 'Properties Owned', 'Risk Weighting','Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df, x=predictor)
```

## Bivariant Analysis

```python
for i, predictor in enumerate(df[['BRId', 'IAId', 'GenderId', 'Amount of Credit Cards', 'Nationality', 'Occupation','Fee Structure','Loyalty Classification', 'Properties Owned', 'Risk Weighting','Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df, x=predictor, hue='Nationality')
```

# Histoplot of value counts for different Occupation

```python
for col in categorical_cols:
  if col == 'Occupation':
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel('Count')
  plt.show()
```

# Numerical Analysis

```python
numerical_col = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univarient analysis and visualization
plt.figure(figsize=(16,10))
for i, col in enumerate(numerical_col):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde = True)
  plt.title(f'Histogram of {col}')
  plt.show()
```

## HeatMaps

```python
numerical_col = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix = df[numerical_col].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

## Insights of EDA:

1. The strongest positive correlation occurs amoung the "Bank Deposits" with "Checking Saving Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintaince the high balances in one account type often holds substantial amount/funds across other accounts as well.
